```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv(
    "../data/raw/u.item",
    sep="|",
    encoding="latin-1",
    header=None
)

genre_columns = [
    "unknown","Action","Adventure","Animation","Children","Comedy",
    "Crime","Documentary","Drama","Fantasy","Film-Noir","Horror",
    "Musical","Mystery","Romance","Sci-Fi","Thriller","War","Western"
]

movies.columns = ["movie_id", "title", "release_date", "video_release",
                  "imdb_url"] + genre_columns

movies = movies[["title"] + genre_columns]
movies.head()
```

```python
from src.content_based import build_genre_matrix

genre_matrix = build_genre_matrix(movies)
similarity_matrix = cosine_similarity(genre_matrix)
```

```python
from src.content_based import content_based_recommender

content_based_recommender(
    movie_title="Toy Story (1995)",
    movies_df=movies,
    similarity_matrix=similarity_matrix,
    top_n=10
)
```
